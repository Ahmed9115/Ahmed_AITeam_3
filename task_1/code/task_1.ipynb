```python
x = np.array([50, 60, 80, 100, 120])
y = np.array ([150, 180, 240, 300, 330])
```

```python
x = np.reshape(x, shape=(-1,1))
x
```

```python
ones = np.ones_like(x)
ones
```

```python
x = np.hstack((x , ones ))
x
```

```python
xtrans = np.matrix_transpose(x)
xtrans
```

```python
# first part of the equation :
temp = np.matmul(xtrans , x)
temp
```

```python
# inverse of the previous part :
temp = np.linalg.inv(temp)
print(temp.shape)
temp
```

```python
# second part of the equation :
temp2 = np.matmul(temp , xtrans)
print(temp2.shape)
temp2
```

```python
theta = temp2@y
theta
```

```python
# the expectation :
house_size = np.array([90 , 1])
expectation = theta@house_size
expectation 
```
