```python
gender_count = zara["Gender"].value_counts()
plt.pie(gender_count, labels=gender_count.index, autopct='%1.1f%%')
plt.title("costumers genders distribution")
#! women likes shopping more than men , that's a fact !!
```

```python
sns.barplot(zara["season"]);
plt.title("season's shopping distribution ");
# people like to shop more on spring !!
```

PLOTTING TIME !!

```python
counts = zara['Product Position'].value_counts()
plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title('Product Position Distribution')
plt.show()
#  the aisle products in zara are much more than that of end-cap (the highlighted ones) !!
```

```python
sns.kdeplot(data = zara ,x =  "price" , hue="Gender")
plt.title("distribution of the prices across genders ")
# here the women colthes are much more ordered than men but in price wise they are within the same range !!
```

```python
sns.kdeplot(data = zara , x = "price", hue = "terms")
plt.title("terms price distribution")
#! the most bought term is the jackets and the least is jeans !!!!!!!!
#! they are within the same price ranges !
```

```python
sns.histplot(data=zara, x="material", hue="season", multiple='stack') # the multiple option makes the colors overlap with wrong indication !!!
plt.xticks(rotation=45, ha='right')   # if i didnt use this , it will make words overlap !!
plt.tight_layout()
plt.title("materials distribution across the seasons")
# a very important one to see what is the most use material in each season !!
# silk is rarely bought ! same for satin  .
```

```python
custom_palette = sns.color_palette("husl", zara["origin"].nunique())

sns.histplot(data=zara, x="origin", hue="material", multiple='stack', palette=custom_palette)
plt.xticks(rotation=45, ha='right')   # if i didnt use this , it will make words overlap !!
plt.tight_layout()
plt.title("countries used materials")
plt.show()
# a very important one also to indicate which country is exporting each material  !!
# the only problem is that there are some materials with the same color so i cant differ which is which and i didnt find a solution !!
```

```python
top_products = zara.groupby('name')['Sales Volume'].sum().nlargest(10)
top_products.plot(kind='bar')
plt.title('Top 10 Products by Sales Volume')
plt.xlabel('Product Name')
plt.ylabel('Sales Volume')
plt.xticks(range(len(top_products)), top_products.index, rotation=45, ha='right')
plt.tight_layout()
plt.show()
#! here i wanted to indicate the amount of the products zara offers , but i couldn't do it myself + the best graph to implement it so i asked ai to help me with this one !!!
```

```python
price_material_seasonal = zara.pivot_table(
    index="material",
    columns="season",
    values="price",)

sns.heatmap(
    price_material_seasonal,
    annot=True,
    fmt=".2f",
    cmap="YlGnBu",
    linewidths=0.5
)
plt.title("relation between Price , Material and Season")
plt.tight_layout()
# i tried multiple time to do pivot table but in each it fails , not sure why so i used the help of ai here also !!
```
