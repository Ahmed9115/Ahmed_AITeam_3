```python
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np
```

 i searched online about the tips dataset from seaborn to learn more about the data to start the process of visualization !

```python
tips_df = sns.load_dataset('tips')
```

```python
tips_df.head()
```

```python
tips_df.info()
```

```python
sns.set_style("darkgrid") 
sns.lineplot(data=tips_df, x="day", y="smoker", marker="*", linestyle="-", hue="sex", errorbar=None, linewidth=2 ,palette="Set1") # note here the errorbar is none because it creates a shadow like area aroung the lines !!!
plt.title("smoking trend")
plt.show()
#! here this is an important one for me cause it tells me that friday is the most smoked one , prehaps it is their's work free day !!
```

```python
sns.set_style("darkgrid") 
sns.lineplot(data=tips_df, x="day", y="total_bill", marker="*", linestyle="-", hue="sex",errorbar=None , linewidth = 3 , palette="cubehelix")
plt.title("bills trend")
plt.show()
#! intersting !! although for females friday for them is the least paid bill but saturday is ! , but for males it increase day by day !!
```

```python
sns.set_style("darkgrid") 
sns.boxplot(data=tips_df, x="size", y = "total_bill" , hue = "smoker")
plt.title("bills chart per groups")
plt.show()
#! here we can see that the smokers are more likely to pay higher bills than non smokers and also the more the size of the group the higher the bill is for non smokers but for smokers it suddenly decreases and i am not sure why !
```

```python
tips_copy = tips_df.copy().pivot_table(index="day", columns="size", values="tip", aggfunc="mean")
tips_copy.head()
#?   dont want to affectt the real data !!
```

```python
sns.heatmap(tips_copy , annot=True  , cmap="YlGnBu")
plt.title("tips accumulation")
plt.show()
#! here the tips increases as the size of the group increases  !!, but it is not applied on saturday !!
```

```python
sns.kdeplot(data=tips_df, x="tip", hue="sex")
plt.title("tips distribution")
plt.show()  
#! well i guess male are more generous than females !!
```

```python
sns.ecdfplot(data=tips_df, x="tip", hue="time")
plt.title("tips distribution per time")
plt.show()
#! there is no much difference but for dinneer time the tips are more than lunch time especially after 6 tips !!
```

```python
sns.violinplot(data=tips_df, x="day", y="total_bill", hue="sex" , split=True)
plt.title("bills distribution to expose outliers")
plt.show()
#! there are billes that  goes beyond 50 dollars especially for saturday and sunday , while the rest has of average between 10 and 20 dollars !!
```
