```python
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression 
from equations import regresion
```

```python
reg = regresion()
```

```python
data = pd.read_csv("California_Houses.csv" )
```

```python
y = data["Median_House_Value"]
y= np.reshape(y , (y.shape[0] ,1 )) # without this line , it is treated as a line of data not a column !
```

```python
data = data.drop("Median_House_Value" , axis = 1)
```

```python
data
```

```python
data.shape
```

```python
x_train,x_tenp_test,y_train,y_temp_test = train_test_split(data , y , test_size=0.3 , random_state=11)
```

```python
x_valid ,x_test,y_valid,y_test = train_test_split(x_tenp_test,y_temp_test , test_size=0.5  ,  random_state=11)
#!here i splited the data into 70% train then 15% valid and testing !!!
```

```python
x_train,x_valid ,x_test , y_train,y_valid ,y_test = reg.normalization(x_train,x_valid ,x_test , y_train,y_valid ,y_test)


x_train = np.hstack([np.ones((x_train.shape[0],1)) , x_train])
x_valid = np.hstack([np.ones((x_valid.shape[0],1)) , x_valid])
x_test = np.hstack([np.ones((x_test.shape[0],1)) , x_test])


```

```python
x_test.shape
```

```python
x_valid.shape
```

linear regression "noraml equation"

```python
final = reg.normal_equation(x_train,y_train)
final.shape
```

#! testing part 

```python
y_validating = x_valid@final
y_testing = x_test@final

print(f"validation score:{r2_score(y_valid,y_validating)}\n")
print(f"testing score:{r2_score(y_test,y_testing)}\n")

print("MSE:")
reg.MSE(y_testing,y_test ,random=False)
print("\nMAE:")
reg.MAE(y_testing,y_test,random=False)

#? i wrote down that score thing based on a video on youtube otherwise i couldn't have compared the results
```

gradient_descent part !!

```python
optimum_weights = reg.gradient_descent_with_mean_squared(x_train,y_train,1000000,0.5)
#lower the iterations if it takes so much time !!
```

```python
grad_valid = x_valid@optimum_weights
grad_test = x_test@optimum_weights
print(f"validation score:{r2_score(y_valid,grad_valid)}")
print(f"testing score:{r2_score(y_test,grad_test)}")
```

```python
print("### error before grag : \n")
print("MSE:")
reg.MSE(x_test,y_test)
print("\nMAE:")
reg.MAE(x_test,y_test)

#! ^^ if u run it several times , you will get different values !! ^^

print("\n### error after grad_squared : \n")
print("MSE:")
reg.MSE(grad_test,y_test ,random=False)
print("\nMAE:")
reg.MAE(grad_test,y_test,random=False)



#! almost as perfect as the normal equation
```

```python
optimum_weights_absolute = reg.gradient_descent_with_mean_absolute(x_train,y_train,1000,0.2) 
```

```python
grad_valid_absolute = x_valid@optimum_weights_absolute
grad_test_absolute = x_test@optimum_weights_absolute
print(f"validation score:{r2_score(y_valid,grad_valid_absolute)}")
print(f"testing score:{r2_score(y_test,grad_test_absolute)}")
```
